```python
"""
=============================================================================
COMPREHENSIVE EDA: PURCHASING POWER PREDICTION SYSTEM
Dataset: 1,262 columns √ó 5,826 rows (2010-2025)
Focus: Which assets preserve purchasing power against inflation
=============================================================================
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# Set plotting style
sns.set_style('whitegrid')
plt.rcParams['figure.dpi'] = 100
plt.rcParams['savefig.dpi'] = 300
plt.rcParams['figure.figsize'] = (15, 6)

# ============================================================================
# 1. LOAD DATA
# ============================================================================

print("="*80)
print("        PURCHASING POWER PREDICTION SYSTEM - EDA")
print("="*80)
print("\nüìÇ Loading dataset...")

# Upload file (uncomment if needed)
# from google.colab import files
# uploaded = files.upload()

# Load dataset
df = pd.read_csv('final_consolidated_dataset.csv')
df['Date'] = pd.to_datetime(df['Date'])

print("‚úÖ Data loaded successfully!")

# ============================================================================
# 2. BASIC INFORMATION
# ============================================================================

print("\n" + "="*80)
print("üìä DATASET OVERVIEW")
print("="*80)

print(f"\nüî¢ Shape: {df.shape[0]:,} rows √ó {df.shape[1]:,} columns")
print(f"üíæ Memory Usage: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
print(f"üìÖ Date Range: {df['Date'].min().strftime('%Y-%m-%d')} to {df['Date'].max().strftime('%Y-%m-%d')}")
print(f"‚è±Ô∏è  Duration: {(df['Date'].max() - df['Date'].min()).days / 365.25:.1f} years")

# Data types
print(f"\nüìã Column Types:")
dtype_counts = df.dtypes.value_counts()
for dtype, count in dtype_counts.items():
    print(f"  {dtype}: {count} columns")

# Display first few rows
print("\nüìÑ First 3 rows:")
display(df.head(3))

# ============================================================================
# 3. MISSING VALUES ANALYSIS
# ============================================================================

print("\n" + "="*80)
print("üîç MISSING VALUES ANALYSIS")
print("="*80)

missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)

missing_df = pd.DataFrame({
    'Column': missing.index,
    'Missing_Count': missing.values,
    'Missing_Percentage': missing_pct.values
})
missing_df = missing_df[missing_df['Missing_Count'] > 0].sort_values('Missing_Count', ascending=False)

print(f"\nüìâ Total columns with missing values: {len(missing_df)}/{len(df.columns)}")
print(f"üìâ Columns with >50% missing: {len(missing_df[missing_df['Missing_Percentage'] > 50])}")

if len(missing_df) > 0:
    print("\nüîù Top 20 Columns with Most Missing Values:")
    print(missing_df.head(20).to_string(index=False))

    # Visualize missing data
    top_missing = missing_df.head(15)

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 6))

    # Bar chart
    ax1.barh(range(len(top_missing)), top_missing['Missing_Percentage'], color='coral')
    ax1.set_yticks(range(len(top_missing)))
    ax1.set_yticklabels(top_missing['Column'], fontsize=10)
    ax1.set_xlabel('Missing Percentage (%)', fontsize=12)
    ax1.set_title('Top 15 Columns with Missing Values', fontsize=14, fontweight='bold')
    ax1.invert_yaxis()
    ax1.grid(axis='x', alpha=0.3)

    # Missing value heatmap (by category)
    crypto_missing = missing_df[missing_df['Column'].str.contains('Bitcoin|Ethereum|Litecoin|Cardano|Solana')]
    traditional_missing = missing_df[missing_df['Column'].str.contains('Gold|Silver|SP500|Oil')]

    categories = ['Crypto Assets', 'Traditional Assets', 'Other']
    counts = [len(crypto_missing), len(traditional_missing),
              len(missing_df) - len(crypto_missing) - len(traditional_missing)]

    ax2.pie(counts, labels=categories, autopct='%1.1f%%', startangle=90,
            colors=['#ff9999','#66b3ff','#99ff99'])
    ax2.set_title('Missing Values by Asset Category', fontsize=14, fontweight='bold')

    plt.tight_layout()
    plt.show()

# ============================================================================
# 4. ECONOMIC INDICATORS ANALYSIS
# ============================================================================

print("\n" + "="*80)
print("üíπ ECONOMIC INDICATORS")
print("="*80)

# Latest values
econ_indicators = {
    'CPI': 'CPI',
    'Inflation Rate': 'Inflation_Rate_YoY',
    'Fed Funds Rate': 'Fed_Funds_Rate',
    'Unemployment': 'Unemployment',
    'Treasury 10Y': 'Treasury_10Y',
    'M2 Growth': 'M2_Growth_Rate'
}

print("\nüìç CURRENT VALUES (Latest Data Point):")
for name, col in econ_indicators.items():
    if col in df.columns:
        latest = df[col].dropna().iloc[-1]
        avg = df[col].dropna().mean()
        print(f"  {name:20s}: {latest:8.2f}  (Avg: {avg:.2f})")

# Inflation analysis
if 'Inflation_Rate_YoY' in df.columns:
    print("\nüìä INFLATION STATISTICS:")
    inflation = df['Inflation_Rate_YoY'].dropna()
    print(f"  Average:   {inflation.mean():.2f}%")
    print(f"  Median:    {inflation.median():.2f}%")
    print(f"  Min:       {inflation.min():.2f}%")
    print(f"  Max:       {inflation.max():.2f}%")
    print(f"  Std Dev:   {inflation.std():.2f}%")

    # Inflation periods
    high_inf = (inflation > 5.0).sum()
    low_inf = (inflation < 2.0).sum()
    print(f"\n  High Inflation (>5%): {high_inf:,} days ({high_inf/len(inflation)*100:.1f}%)")
    print(f"  Low Inflation (<2%):  {low_inf:,} days ({low_inf/len(inflation)*100:.1f}%)")

    # Plot inflation over time
    fig, axes = plt.subplots(2, 1, figsize=(16, 10))

    # Time series
    axes[0].plot(df['Date'], df['Inflation_Rate_YoY'], linewidth=2, color='darkblue', label='Inflation Rate')
    axes[0].axhline(2.0, color='green', linestyle='--', linewidth=2, label='Fed Target (2%)')
    axes[0].axhline(5.0, color='red', linestyle='--', linewidth=2, label='High Inflation (5%)')
    axes[0].fill_between(df['Date'], 0, df['Inflation_Rate_YoY'], alpha=0.3)
    axes[0].set_title('Inflation Rate Over Time (2010-2025)', fontsize=16, fontweight='bold')
    axes[0].set_ylabel('Inflation Rate (%)', fontsize=12)
    axes[0].legend(loc='upper right')
    axes[0].grid(alpha=0.3)

    # Distribution
    axes[1].hist(inflation, bins=50, edgecolor='black', alpha=0.7, color='steelblue')
    axes[1].axvline(inflation.mean(), color='red', linestyle='--', linewidth=2, label=f'Mean: {inflation.mean():.2f}%')
    axes[1].axvline(inflation.median(), color='green', linestyle='--', linewidth=2, label=f'Median: {inflation.median():.2f}%')
    axes[1].set_title('Inflation Rate Distribution', fontsize=16, fontweight='bold')
    axes[1].set_xlabel('Inflation Rate (%)', fontsize=12)
    axes[1].set_ylabel('Frequency', fontsize=12)
    axes[1].legend()
    axes[1].grid(alpha=0.3)

    plt.tight_layout()
    plt.show()

# ============================================================================
# 5. ASSET PRICES ANALYSIS
# ============================================================================

print("\n" + "="*80)
print("üí∞ ASSET PRICES")
print("="*80)

assets_prices = {
    'Bitcoin': 'Bitcoin_Price',
    'Ethereum': 'Ethereum_Price',
    'Litecoin': 'Litecoin_Price',
    'Gold': 'Gold',
    'Silver': 'Silver',
    'SP500': 'SP500',
    'NASDAQ': 'NASDAQ',
    'Oil': 'Oil'
}

print("\nüìç CURRENT PRICES (Latest):")
for name, col in assets_prices.items():
    if col in df.columns:
        latest = df[col].dropna().iloc[-1]
        first = df[col].dropna().iloc[0]
        growth = ((latest - first) / first * 100)
        print(f"  {name:15s}: ${latest:12,.2f}  (Growth since 2010: {growth:+.1f}%)")

# Plot asset prices
fig, axes = plt.subplots(2, 2, figsize=(18, 12))

# Bitcoin
if 'Bitcoin_Price' in df.columns:
    axes[0, 0].plot(df['Date'], df['Bitcoin_Price'], linewidth=2, color='orange')
    axes[0, 0].set_title('Bitcoin Price Over Time', fontsize=14, fontweight='bold')
    axes[0, 0].set_ylabel('Price (USD)', fontsize=12)
    axes[0, 0].set_yscale('log')
    axes[0, 0].grid(alpha=0.3)

# Gold
if 'Gold' in df.columns:
    axes[0, 1].plot(df['Date'], df['Gold'], linewidth=2, color='gold')
    axes[0, 1].set_title('Gold Price Over Time', fontsize=14, fontweight='bold')
    axes[0, 1].set_ylabel('Price (USD/oz)', fontsize=12)
    axes[0, 1].grid(alpha=0.3)

# SP500
if 'SP500' in df.columns:
    axes[1, 0].plot(df['Date'], df['SP500'], linewidth=2, color='darkgreen')
    axes[1, 0].set_title('S&P 500 Index Over Time', fontsize=14, fontweight='bold')
    axes[1, 0].set_ylabel('Index Level', fontsize=12)
    axes[1, 0].grid(alpha=0.3)

# Ethereum
if 'Ethereum_Price' in df.columns:
    axes[1, 1].plot(df['Date'], df['Ethereum_Price'], linewidth=2, color='purple')
    axes[1, 1].set_title('Ethereum Price Over Time', fontsize=14, fontweight='bold')
    axes[1, 1].set_ylabel('Price (USD)', fontsize=12)
    axes[1, 1].set_yscale('log')
    axes[1, 1].grid(alpha=0.3)

plt.tight_layout()
plt.show()

# ============================================================================
# 6. PURCHASING POWER MULTIPLIER ANALYSIS (CORE)
# ============================================================================

print("\n" + "="*80)
print("üéØ PURCHASING POWER MULTIPLIER ANALYSIS (5Y)")
print("="*80)

# Find all 5Y PP multiplier columns
pp_5y_cols = [col for col in df.columns if 'PP_Multiplier_5Y' in col and 'Cash' not in col]

if len(pp_5y_cols) > 0:
    print(f"\nüìä Found {len(pp_5y_cols)} assets with 5Y PP Multiplier data\n")

    # Calculate statistics
    pp_stats = []
    for col in pp_5y_cols:
        asset = col.replace('_PP_Multiplier_5Y', '')
        data = df[col].dropna()

        if len(data) > 100:  # Only if sufficient data
            pp_stats.append({
                'Asset': asset,
                'Mean': data.mean(),
                'Median': data.median(),
                'Std_Dev': data.std(),
                'Min': data.min(),
                'Max': data.max(),
                'Data_Points': len(data),
                'Above_1x': (data > 1.0).sum() / len(data) * 100
            })

    pp_df = pd.DataFrame(pp_stats).sort_values('Mean', ascending=False)

    print("üèÜ TOP 20 ASSETS BY AVERAGE 5Y PP MULTIPLIER:")
    print(pp_df.head(20)[['Asset', 'Mean', 'Median', 'Above_1x']].to_string(index=False))

    # Visualizations
    fig, axes = plt.subplots(2, 2, figsize=(18, 12))

    # 1. Bar chart of top assets
    top_15 = pp_df.head(15)
    bars = axes[0, 0].barh(range(len(top_15)), top_15['Mean'], edgecolor='black', linewidth=1.5)

    # Color by performance
    for i, (bar, mean) in enumerate(zip(bars, top_15['Mean'])):
        if mean > 1.5:
            bar.set_color('darkgreen')
        elif mean > 1.0:
            bar.set_color('lightgreen')
        elif mean > 0.9:
            bar.set_color('orange')
        else:
            bar.set_color('red')

    axes[0, 0].set_yticks(range(len(top_15)))
    axes[0, 0].set_yticklabels(top_15['Asset'], fontsize=10)
    axes[0, 0].axvline(1.0, color='red', linestyle='--', linewidth=2, label='Break-even')
    axes[0, 0].set_xlabel('Average PP Multiplier', fontsize=12)
    axes[0, 0].set_title('Top 15 Assets by 5Y PP Multiplier', fontsize=14, fontweight='bold')
    axes[0, 0].invert_yaxis()
    axes[0, 0].legend()
    axes[0, 0].grid(axis='x', alpha=0.3)

    # 2. Distribution of top 4 assets
    top_4_assets = pp_df.head(4)['Asset'].tolist()
    for asset in top_4_assets:
        col = f'{asset}_PP_Multiplier_5Y'
        if col in df.columns:
            data = df[col].dropna()
            axes[0, 1].hist(data, bins=50, alpha=0.5, label=asset, edgecolor='black')

    axes[0, 1].axvline(1.0, color='red', linestyle='--', linewidth=2)
    axes[0, 1].set_xlabel('PP Multiplier', fontsize=12)
    axes[0, 1].set_ylabel('Frequency', fontsize=12)
    axes[0, 1].set_title('PP Multiplier Distribution (Top 4 Assets)', fontsize=14, fontweight='bold')
    axes[0, 1].legend()
    axes[0, 1].grid(alpha=0.3)

    # 3. Time series of top 3 assets
    top_3_assets = pp_df.head(3)['Asset'].tolist()
    for asset in top_3_assets:
        col = f'{asset}_PP_Multiplier_5Y'
        if col in df.columns:
            axes[1, 0].plot(df['Date'], df[col], linewidth=2, label=asset, alpha=0.8)

    axes[1, 0].axhline(1.0, color='red', linestyle='--', linewidth=2, label='Break-even')
    axes[1, 0].set_ylabel('PP Multiplier', fontsize=12)
    axes[1, 0].set_title('PP Multiplier Over Time (Top 3)', fontsize=14, fontweight='bold')
    axes[1, 0].legend()
    axes[1, 0].grid(alpha=0.3)

    # 4. Win rate (% above 1.0)
    win_rate_data = pp_df.head(15)
    axes[1, 1].barh(range(len(win_rate_data)), win_rate_data['Above_1x'],
                    edgecolor='black', color='steelblue')
    axes[1, 1].set_yticks(range(len(win_rate_data)))
    axes[1, 1].set_yticklabels(win_rate_data['Asset'], fontsize=10)
    axes[1, 1].set_xlabel('% of Time Above 1.0x', fontsize=12)
    axes[1, 1].set_title('PP Preservation Success Rate', fontsize=14, fontweight='bold')
    axes[1, 1].invert_yaxis()
    axes[1, 1].grid(axis='x', alpha=0.3)

    plt.tight_layout()
    plt.show()

    # Key insights
    print("\nüí° KEY INSIGHTS:")
    best_asset = pp_df.iloc[0]
    print(f"  ü•á Best Asset: {best_asset['Asset']} (Avg: {best_asset['Mean']:.3f}x)")
    print(f"  üìà Preserves PP {best_asset['Above_1x']:.1f}% of the time")

    worst_asset = pp_df.iloc[-1]
    print(f"  ‚ö†Ô∏è  Worst Asset: {worst_asset['Asset']} (Avg: {worst_asset['Mean']:.3f}x)")

# ============================================================================
# 7. BEST ASSET ANALYSIS
# ============================================================================

print("\n" + "="*80)
print("üèÜ BEST ASSET DISTRIBUTION")
print("="*80)

if 'Best_Asset_5Y' in df.columns:
    best_asset_counts = df['Best_Asset_5Y'].value_counts()

    print("\nüìä Which asset wins most often (Best_Asset_5Y)?")
    print(best_asset_counts.head(10).to_string())

    # Visualize
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 6))

    # Bar chart
    best_asset_counts.head(10).plot(kind='bar', ax=ax1, color='steelblue', edgecolor='black')
    ax1.set_title('Best Asset Distribution (5Y Horizon)', fontsize=16, fontweight='bold')
    ax1.set_ylabel('Number of Days', fontsize=12)
    ax1.set_xlabel('Asset', fontsize=12)
    ax1.tick_params(axis='x', rotation=45)
    ax1.grid(axis='y', alpha=0.3)

    # Pie chart
    top_5 = best_asset_counts.head(5)
    others = best_asset_counts.iloc[5:].sum()
    pie_data = list(top_5.values) + [others]
    pie_labels = list(top_5.index) + ['Others']

    ax2.pie(pie_data, labels=pie_labels, autopct='%1.1f%%', startangle=90)
    ax2.set_title('Best Asset Market Share', fontsize=16, fontweight='bold')

    plt.tight_layout()
    plt.show()

    # Winner percentage
    total_days = len(df['Best_Asset_5Y'].dropna())
    winner = best_asset_counts.index[0]
    winner_pct = best_asset_counts.iloc[0] / total_days * 100
    print(f"\nüéØ {winner} is the best asset {winner_pct:.1f}% of the time")

# ============================================================================
# 8. VOLATILITY ANALYSIS
# ============================================================================

print("\n" + "="*80)
print("üìä VOLATILITY ANALYSIS")
print("="*80)

vol_cols = [col for col in df.columns if 'Volatility_90D' in col]

if len(vol_cols) > 0:
    print(f"\nüìà Comparing volatility across {len(vol_cols)} assets:\n")

    vol_data = []
    for col in vol_cols:
        asset = col.replace('_Volatility_90D', '')
        data = df[col].dropna()
        if len(data) > 100:
            vol_data.append({
                'Asset': asset,
                'Avg_Volatility': data.mean(),
                'Risk_Tier': 'Low' if data.mean() < 15 else ('Medium' if data.mean() < 30 else 'High')
            })

    vol_df = pd.DataFrame(vol_data).sort_values('Avg_Volatility')

    print("üîù LEAST VOLATILE (Safest):")
    print(vol_df.head(10).to_string(index=False))

    print("\n‚ö†Ô∏è  MOST VOLATILE (Riskiest):")
    print(vol_df.tail(10).to_string(index=False))

    # Visualize
    plt.figure(figsize=(14, 8))
    bars = plt.barh(range(len(vol_df)), vol_df['Avg_Volatility'], edgecolor='black')

    # Color by risk
    for i, (bar, vol) in enumerate(zip(bars, vol_df['Avg_Volatility'])):
        if vol < 15:
            bar.set_color('green')
        elif vol < 30:
            bar.set_color('yellow')
        elif vol < 60:
            bar.set_color('orange')
        else:
            bar.set_color('red')

    plt.yticks(range(len(vol_df)), vol_df['Asset'], fontsize=10)
    plt.xlabel('Average 90-Day Volatility (%)', fontsize=12)
    plt.title('Asset Volatility Comparison', fontsize=16, fontweight='bold')
    plt.axvline(15, color='green', linestyle='--', alpha=0.5, label='Low Risk')
    plt.axvline(30, color='orange', linestyle='--', alpha=0.5, label='Medium Risk')
    plt.axvline(60, color='red', linestyle='--', alpha=0.5, label='High Risk')
    plt.legend()
    plt.grid(axis='x', alpha=0.3)
    plt.tight_layout()
    plt.show()

# ============================================================================
# 9. CORRELATION ANALYSIS
# ============================================================================

print("\n" + "="*80)
print("üîó CORRELATION ANALYSIS")
print("="*80)

# Select key PP multiplier columns
key_pp_cols = []
for asset in ['Bitcoin', 'Ethereum', 'Gold', 'Silver', 'SP500', 'NASDAQ',
              'Oil', 'Apple', 'Microsoft']:
    col = f'{asset}_PP_Multiplier_5Y'
    if col in df.columns:
        key_pp_cols.append(col)

if len(key_pp_cols) >= 3:
    # Create simplified names
    simplified_names = {col: col.replace('_PP_Multiplier_5Y', '') for col in key_pp_cols}

    corr_matrix = df[key_pp_cols].rename(columns=simplified_names).corr()

    plt.figure(figsize=(12, 10))
    sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm',
                center=0, square=True, linewidths=1.5, cbar_kws={"shrink": 0.8})
    plt.title('PP Multiplier Correlations (5Y)', fontsize=16, fontweight='bold')
    plt.tight_layout()
    plt.show()

    # Find highly correlated pairs
    print("\nüîó Highly Correlated Assets (>0.7):")
    for i in range(len(corr_matrix.columns)):
        for j in range(i+1, len(corr_matrix.columns)):
            corr_val = corr_matrix.iloc[i, j]
            if abs(corr_val) > 0.7:
                print(f"  {corr_matrix.columns[i]} ‚Üî {corr_matrix.columns[j]}: {corr_val:.3f}")

# ============================================================================
# 10. REAL COMMODITY PURCHASING POWER
# ============================================================================

print("\n" + "="*80)
print("ü•ö REAL COMMODITY PURCHASING POWER")
print("="*80)

if 'Eggs_Per_100USD' in df.columns:
    latest_eggs = df['Eggs_Per_100USD'].dropna().iloc[-1]
    first_eggs = df['Eggs_Per_100USD'].dropna().iloc[0]

    print(f"\nü•ö EGGS:")
    print(f"  2010: $100 could buy {first_eggs:.1f} dozen eggs")
    print(f"  2025: $100 can buy {latest_eggs:.1f} dozen eggs")
    print(f"  Loss: {(first_eggs - latest_eggs)/first_eggs*100:.1f}%")

if 'Milk_Gallons_Per_100USD' in df.columns:
    latest_milk = df['Milk_Gallons_Per_100USD'].dropna().iloc[-1]
    first_milk = df['Milk_Gallons_Per_100USD'].dropna().iloc[0]

    print(f"\nü•õ MILK:")
    print(f"  2010: $100 could buy {first_milk:.1f} gallons")
    print(f"  2025: $100 can buy {latest_milk:.1f} gallons")
    print(f"  Loss: {(first_milk - latest_milk)/first_milk*100:.1f}%")

# Plot real purchasing power
if 'Eggs_Per_100USD' in df.columns and 'Milk_Gallons_Per_100USD' in df.columns:
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

    ax1.plot(df['Date'], df['Eggs_Per_100USD'], linewidth=2, color='brown', label='Eggs')
    ax1.set_title('Egg Purchasing Power of $100', fontsize=14, fontweight='bold')
    ax1.set_ylabel('Dozen Eggs per $100', fontsize=12)
    ax1.grid(alpha=0.3)

    ax2.plot(df['Date'], df['Milk_Gallons_Per_100USD'], linewidth=2, color='skyblue', label='Milk')
    ax2.set_title('Milk Purchasing Power of $100', fontsize=14, fontweight='bold')
    ax2.set_ylabel('Gallons per $100', fontsize=12)
    ax2.grid(alpha=0.3)

    plt.tight_layout()
    plt.show()

# ============================================================================
# 11. DATA QUALITY SUMMARY
# ============================================================================

print("\n" + "="*80)
print("‚úÖ DATA QUALITY SUMMARY")
print("="*80)

# Duplicates
duplicates = df.duplicated().sum()
print(f"\nüîç Duplicate rows: {duplicates}")

# Complete cases
complete_cases = df.dropna().shape[0]
print(f"üìã Complete cases (no missing): {complete_cases:,} ({complete_cases/len(df)*100:.1f}%)")

# Date continuity
df_sorted = df.sort_values('Date')
date_diffs = df_sorted['Date'].diff()
gaps = (date_diffs > pd.Timedelta(days=7)).sum()
print(f"üìÖ Date gaps (>7 days): {gaps}")

# Value ranges
print(f"\nüìä Value Ranges:")
print(f"  Inflation: {df['Inflation_Rate_YoY'].min():.2f}% to {df['Inflation_Rate_YoY'].max():.2f}%")
if 'Bitcoin_Price' in df.columns:
    print(f"  Bitcoin:   ${df['Bitcoin_Price'].min():.2f} to ${df['Bitcoin_Price'].max():,.2f}")
if 'Gold' in df.columns:
    print(f"  Gold:      ${df['Gold'].min():.2f} to ${df['Gold'].max():.2f}")

# ============================================================================
# 12. FINAL SUMMARY & RECOMMENDATIONS
# ============================================================================

print("\n" + "="*80)
print("üìù EDA SUMMARY & NEXT STEPS")
print("="*80)

print("\n‚úÖ DATASET QUALITY:")
print(f"  ‚Ä¢ {df.shape[0]:,} rows √ó {df.shape[1]:,} columns")
print(f"  ‚Ä¢ {(df['Date'].max() - df['Date'].min()).days / 365.25:.1f} years of data")
print(f"  ‚Ä¢ {len(missing_df[missing_df['Missing_Percentage'] > 50])} columns with >50% missing")
print(f"  ‚Ä¢ {duplicates} duplicate rows")

if len(pp_5y_cols) > 0:
    print(f"\nüéØ KEY FINDINGS:")
    print(f"  ‚Ä¢ Best PP Asset: {pp_df.iloc[0]['Asset']} ({pp_df.iloc[0]['Mean']:.3f}x avg multiplier)")
    print(f"  ‚Ä¢ Most Volatile: {vol_df.iloc[-1]['Asset']} ({vol_df.iloc[-1]['Avg_Volatility']:.1f}% volatility)")
    print(f"  ‚Ä¢ Safest: {vol_df.iloc[0]['Asset']} ({vol_df.iloc[0]['Avg_Volatility']:.1f}% volatility)")

print(f"\nüöÄ NEXT STEPS:")
print("  1. Handle missing values (forward-fill, drop columns >50% missing)")
print("  2. Create train/test split (time-based, 80/20)")
print("  3. Feature selection (identify top predictors)")
print("  4. Build baseline models (6 ML tasks)")
print("  5. Train advanced models (XGBoost, LSTM)")

print("\n" + "="*80)
print("‚úÖ EDA COMPLETE!")
print("="*80)

# Save summary statistics
summary_stats = {
    'total_rows': len(df),
    'total_columns': len(df.columns),
    'date_range': f"{df['Date'].min()} to {df['Date'].max()}",
    'duration_years': (df['Date'].max() - df['Date'].min()).days / 365.25,
    'missing_columns': len(missing_df),
    'duplicate_rows': duplicates,
    'avg_inflation': df['Inflation_Rate_YoY'].mean() if 'Inflation_Rate_YoY' in df.columns else None
}

print("\nüíæ Summary saved to memory. Ready for next steps!")
```
